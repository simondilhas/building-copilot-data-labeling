```python
def add_classifications_to_new_project(df_new_project: pd.DataFrame, df_mapping: pd.DataFrame, df_classification: pd.DataFrame) -> pd.DataFrame:

  df_mapped = df_new_project.merge(df_mapping, on='OrigGuids', how='left')
  df_mapped = df_mapped.drop_duplicates()

  df_enriched_with_classifications = df_mapped.merge(df_classification, on='CopilotClassification', how='left')
  df_enriched_with_classifications = df_enriched_with_classifications.drop_duplicates()
  df_enriched_with_classifications.head()

  return(df_enriched_with_classifications)

#df_mapped = add_classifications_to_new_project(df_new_project, df_mapping, df_classification)
#df_mapped.head()
```

```python
def split_column_at_comma(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    split_values = df[column_name].str.split(',', expand=True)
    new_column_names = [f"{column_name}_part{i+1}" for i in range(split_values.shape[1])]
    split_values.columns = new_column_names
    df = pd.concat([df, split_values], axis=1)

    return df
```

```python
def merge_enriched_data(df_original, df_enriched, key_column, column_to_enrich):
    """
    Merge the original dataframe with the enriched dataframe, overriding original data only when empty.
    Used together with function "check_spacial_relationship".

    Parameters:
    df_original (DataFrame): Original dataframe.
    df_enriched (DataFrame): Enriched dataframe.
    key_column (str): Name of the common key column.
    column_to_enrich (str): Name of the column to be enriched.

    Returns:
    DataFrame: Merged dataframe.
    """
    # Merge the dataframes on 'Guid' using left join
    merged_df = pd.merge(df_original, df_enriched, on=key_column, how='left', suffixes=('', '_enriched'))
    #merged_df = pd.merge(df_original, df_enriched, left_on=key_column, right_on=sec_key_optional, how='left', suffixes=('', '_enriched'))

    # Override the original data only when it's empty
    merged_df[column_to_enrich] = merged_df.apply(lambda row: row[f"{column_to_enrich}_enriched"] if pd.isnull(row[column_to_enrich]) else row[column_to_enrich], axis=1)

    # Drop the enriched column
    merged_df.drop(columns=[f"{column_to_enrich}_enriched"], inplace=True)

    #TODO why is this necessary?
    #merged_df = merged_df.drop_duplicates(subset=['Guid'])

    return merged_df
```

```python
def write_above_ground_to_elements(df: pd.DataFrame) -> pd.DataFrame:

  df_space = df[(df['IfcElement'] == 'SPACE') & (df['PredefinedType'] != 'GFA')]
  df_space = df_space[['Guid', 'AboveGround']]

  for i in range(len(df)):
      if ((df_space['Guid'] == df.loc[i, 'RelationSpaceGUID_part1']) | (df_space['Guid'] == df.loc[i, 'RelationSpaceGUID_part2'])).any():
          df.loc[i, 'AboveGround'] = df_space.loc[df_space['Guid'] == df.loc[i, 'RelationSpaceGUID_part1'], 'AboveGround'].iloc[0]
      else:
          pass

  return(df)
```

```python
#TODO above/below ground for GFA works for flat sites, find a better way for buildings in a slope
#e.g. % of the spaces in the floor above/Below ground

ground = -0.1

def above_ground(df: pd.DataFrame, pre_selection) -> pd.DataFrame:
    above_ground_df = (df['AboveGround'] == True) & pre_selection
    return above_ground_df

def below_ground(df: pd.DataFrame, pre_selection) -> pd.DataFrame:
    below_ground_df = (df['AboveGround'] == False) & pre_selection
    return below_ground_df

def above_ground_gfa(df: pd.DataFrame, pre_selection) -> pd.DataFrame:
    above_ground_df = (df['Z'] >= ground) & pre_selection
    return above_ground_df

def below_ground_gfa(df: pd.DataFrame, pre_selection) -> pd.DataFrame:
    below_ground_df = (df['Z'] < ground) & pre_selection
    return below_ground_df


```

```python
def calculate_sia_area_metrics(df: pd.DataFrame) -> pd.DataFrame:

    'Given a raw DataFrame (compliant with our format), returns the main SIA metrics.'

    spaces_df = df[df['IfcElement'] == 'SPACE'].copy()

    is_void_space = spaces_df['CopilotClassification'] == '99.01.00 Luftraum'
    is_void_space_above_ground = above_ground(spaces_df, is_void_space)
    is_void_space_below_ground = below_ground(spaces_df, is_void_space)

    is_gfa = spaces_df['PredefinedType'] == 'GFA'
    is_gfa_above_ground = above_ground_gfa(spaces_df, is_gfa)
    is_gfa_below_ground = below_ground_gfa(spaces_df, is_gfa)

    is_ngf = spaces_df['SIA416-Level1'] == 'NGF'
    is_ngf_above_ground = above_ground(spaces_df, is_ngf)
    is_ngf_below_ground = above_ground(spaces_df, is_ngf)

    #TODO: Check why someSIA_HNV_BELOW_GROUNDic mistake

    is_nf = spaces_df['SIA416-Level2'] == 'NF'
    is_nf_above_ground = above_ground(spaces_df, is_nf)
    is_nf_below_ground = below_ground(spaces_df, is_nf)

    is_vf = spaces_df['SIA416-Level2'] == 'VF'
    is_vf_above_ground = above_ground(spaces_df, is_vf)
    is_vf_below_ground = below_ground(spaces_df, is_vf)

    is_ff = spaces_df['SIA416-Level2'] == 'FF'
    is_ff_above_ground = above_ground(spaces_df, is_ff)
    is_ff_below_ground = below_ground(spaces_df, is_ff)

    is_internal = spaces_df['PredefinedType'] == 'INTERNAL'
    is_internal_above_ground = above_ground(spaces_df, is_internal)
    is_internal_below_ground = below_ground(spaces_df, is_internal)

    is_hnf = spaces_df['SIA416-Level3'] == 'HNF'
    is_hnf_above_ground = above_ground(spaces_df, is_hnf)
    is_hnf_below_ground = below_ground(spaces_df, is_hnf)

    is_nnf = spaces_df['SIA416-Level3'] == 'NNF'
    is_nnf_above_ground = above_ground(spaces_df, is_nnf)
    is_nnf_below_ground = below_ground(spaces_df, is_nnf)

    is_anf = spaces_df['SIA416-Level3'] == 'ANF'
    is_anf_above_ground = above_ground(spaces_df, is_anf)
    is_anf_below_ground = below_ground(spaces_df, is_anf)

    is_avf = spaces_df['SIA416-Level3'] == 'AVF'
    is_avf_above_ground = above_ground(spaces_df, is_avf)
    is_avf_below_ground = below_ground(spaces_df, is_avf)

    is_buf = spaces_df['SIA416-Level3'] == 'BUF'
    is_buf_above_ground = above_ground(spaces_df, is_buf)
    is_buf_below_ground = below_ground(spaces_df, is_buf)


    # General
    SPACE_INTERNAL_AREA = spaces_df[is_internal]['NetFloorArea'].sum()
    SPACE_INTERNAL_AREA_ABOVE_GROUND = spaces_df[is_internal_above_ground]['NetFloorArea'].sum()
    SPACE_INTERNAL_AREA_BELOW_GROUND = spaces_df[is_internal_below_ground]['NetFloorArea'].sum()

    SIA_ANF=spaces_df[is_anf]['NetFloorArea'].sum()
    SIA_ANF_ABOVE_GROUND=spaces_df[is_anf_above_ground]['NetFloorArea'].sum()
    SIA_ANF_BELOW_GROUND=spaces_df[is_anf_below_ground]['NetFloorArea'].sum()

    SIA_AVF=spaces_df[is_avf]['NetFloorArea'].sum()
    SIA_AVF_ABOVE_GROUND=spaces_df[is_avf_above_ground]['NetFloorArea'].sum()
    SIA_AVF_BELOW_GROUND=spaces_df[is_avf_below_ground]['NetFloorArea'].sum()

    SIA_BUF=spaces_df[is_avf]['NetFloorArea'].sum()
    SIA_BUF_ABOVE_GROUND=spaces_df[is_avf_above_ground]['NetFloorArea'].sum()
    SIA_BUF_BELOW_GROUND=spaces_df[is_avf_below_ground]['NetFloorArea'].sum()

    SIA_FF=spaces_df[is_ff]['NetFloorArea'].sum()
    SIA_FF_ABOVE_GROUND=spaces_df[is_ff_above_ground]['NetFloorArea'].sum()
    SIA_FF_BELOW_GROUND=spaces_df[is_ff_below_ground]['NetFloorArea'].sum()

    SIA_GF=spaces_df[is_gfa]['NetFloorArea'].sum() - spaces_df[is_void_space]['NetFloorArea'].sum()
    SIA_GF_ABOVE_GROUND=spaces_df[is_gfa_above_ground]['NetFloorArea'].sum() - spaces_df[is_void_space_above_ground]['NetFloorArea'].sum()
    SIA_GF_BELOW_GROUND=spaces_df[is_gfa_below_ground]['NetFloorArea'].sum() - spaces_df[is_void_space_below_ground]['NetFloorArea'].sum()

    SIA_HNF=spaces_df[is_hnf]['NetFloorArea'].sum()
    SIA_HNF_ABOVE_GROUND=spaces_df[is_hnf_above_ground]['NetFloorArea'].sum()
    SIA_HNF_BELOW_GROUND=spaces_df[is_hnf_below_ground]['NetFloorArea'].sum()

    SIA_NF=spaces_df[is_nf]['NetFloorArea'].sum()
    SIA_NF_ABOVE_GROUND=spaces_df[is_nf_above_ground]['NetFloorArea'].sum()
    SIA_NF_BELOW_GROUND=spaces_df[is_nf_below_ground]['NetFloorArea'].sum()

    SIA_NGF=spaces_df[is_ngf]['NetFloorArea'].sum()
    SIA_NGF_ABOVE_GROUND=spaces_df[is_ngf_above_ground]['NetFloorArea'].sum()
    SIA_NGF_BELOW_GROUND=spaces_df[is_ngf_below_ground]['NetFloorArea'].sum()


    SIA_NNF=spaces_df[is_nnf]['NetFloorArea'].sum()
    SIA_NNF_ABOVE_GROUND=spaces_df[is_nnf_above_ground]['NetFloorArea'].sum()
    SIA_NNF_BELOW_GROUND=spaces_df[is_nnf_below_ground]['NetFloorArea'].sum()

    SIA_VF=spaces_df[is_vf]['NetFloorArea'].sum()
    SIA_VF_ABOVE_GROUND=spaces_df[is_vf_above_ground]['NetFloorArea'].sum()
    SIA_VF_BELOW_GROUND=spaces_df[is_vf_below_ground]['NetFloorArea'].sum()

    SIA_KF=SIA_GF - SIA_NGF



    sia_dict = {
                'SIA_ANF': [SIA_ANF],
                'SIA_ANF_ABOVE_GROUND': [SIA_ANF_ABOVE_GROUND],
                'SIA_ANF_BELOW_GROUND': [SIA_ANF_BELOW_GROUND],
                'SIA_AVF': [SIA_AVF],
                'SIA_AVF_ABOVE_GROUND': [SIA_AVF_ABOVE_GROUND],
                'SIA_AVF_BELOW_GROUND': [SIA_AVF_BELOW_GROUND],
                'SIA_BUF': [SIA_BUF],
                'SIA_BUF_ABOVE_GROUND': [SIA_BUF_ABOVE_GROUND],
                'SIA_BUF_BELOW_GROUND': [SIA_BUF_BELOW_GROUND],
                'SIA_FF': [SIA_FF],
                'SIA_FF_ABOVE_GROUND': [SIA_FF_ABOVE_GROUND],
                'SIA_FF_BELOW_GROUND': [SIA_FF_BELOW_GROUND],
                'SIA_GF': [SIA_GF],
                'SIA_GF_ABOVE_GROUND': [SIA_GF_ABOVE_GROUND],
                'SIA_GF_BELOW_GROUND': [SIA_GF_BELOW_GROUND],
                'SIA_HNF': [SIA_HNF],
                'SIA_HNF_ABOVE_GROUND': [SIA_HNF_ABOVE_GROUND],
                'SIA_HNF_BELOW_GROUND': [SIA_HNF_BELOW_GROUND],
                'SIA_NF': [SIA_NF],
                'SIA_NF_ABOVE_GROUND': [SIA_NF_ABOVE_GROUND],
                'SIA_NF_BELOW_GROUND': [SIA_NF_BELOW_GROUND],
                'SIA_NNF': [SIA_NNF],
                'SIA_NNF_ABOVE_GROUND': [SIA_NNF_ABOVE_GROUND],
                'SIA_NNF_BELOW_GROUND': [SIA_NNF_BELOW_GROUND],
                'SIA_VF': [SIA_VF],
                'SIA_VF_ABOVE_GROUND': [SIA_VF_ABOVE_GROUND],
                'SIA_VF_BELOW_GROUND': [SIA_VF_BELOW_GROUND],
                'SPACE_INTERNAL_AREA': [SPACE_INTERNAL_AREA],
                'SPACE_INTERNAL_AREA_ABOVE_GROUND': [SPACE_INTERNAL_AREA_ABOVE_GROUND],
                'SPACE_INTERNAL_AREA_BELOW_GROUND': [SPACE_INTERNAL_AREA_BELOW_GROUND],
                'SIA_KF': [SIA_KF]
    }

    df_return = pd.DataFrame(sia_dict)

    return(df_return)

df_sia_area_test = calculate_sia_area_metrics(df_test_data)
df_sia_area_test.head()
```

```python
def general_volume_metrics(df: pd.DataFrame) -> pd.DataFrame:

    'Given a raw DataFrame (compliant with our format), returns the construction volumes.'

    elements_df = df[df['IfcElement'] != 'SPACE'].copy()
    space_df = df[df['IfcElement'] == 'SPACE'].copy()

    is_wall = elements_df['IfcElement'] == 'WALLSTANDARDCASE'
    is_wall_above_ground = above_ground(elements_df, is_wall)
    is_wall_below_ground = below_ground(elements_df, is_wall)

    is_slab = elements_df['IfcElement'] == 'SLAB'

    #TODO How to calculate the above/below ground slabs, (Relationship to Space?)
    is_slab_above_ground = above_ground_gfa(elements_df, is_slab)
    is_slab_below_ground = below_ground_gfa(elements_df, is_slab)

    is_net_space = space_df['PredefinedType'] != 'GFA'
    is_net_space_above_ground = above_ground(elements_df, is_net_space)
    is_net_space_below_ground = below_ground(elements_df, is_net_space)

    #TODO tempory fix of bug: Net to Gross after Ralph fixed the bug
    elements_df['HelpGrossVolume'] = elements_df['NetSideArea'] * elements_df['Width']

    #TODO: Change after Ralph fixes the bug: 'HelpGrossWallVolume' to 'GrossVolume'
    bugfix='HelpGrossVolume'

    COP_KV_WALL=elements_df[is_wall][bugfix].sum()
    COP_KV_SLAB=elements_df[is_slab]['GrossVolume'].sum()
    COP_KV = COP_KV_WALL + COP_KV_SLAB

    COP_KV_WALL_ABOVE_GROUND=elements_df[is_wall_above_ground][bugfix].sum()
    COP_KV_SLAB_ABOVE_GROUND=elements_df[is_slab_above_ground]['GrossVolume'].sum()  #TODO SLABS don't have the above ground attribute
    COP_KV_ABOVE_GROUND = COP_KV_WALL_ABOVE_GROUND + COP_KV_SLAB_ABOVE_GROUND

    COP_KV_WALL_BELOW_GROUND=elements_df[is_wall_below_ground][bugfix].sum()
    COP_KV_SLAB_BELOW_GROUND=elements_df[is_slab_below_ground]['GrossVolume'].sum() #TODO SLABS don't have the above ground attribute
    COP_KV_BELOW_GROUND = COP_KV_WALL_BELOW_GROUND + COP_KV_SLAB_BELOW_GROUND

    #TODO: Check what's a good Volmetric metric and definition (I include exterior spaces with the idea that they provide useable space aswell)

    COP_VOLUME_NET_SPACE=space_df[is_net_space]['NetVolume'].sum()
    COP_VOLUME_NET_SPACE_ABOVE_GROUND=space_df[is_net_space_above_ground]['NetVolume'].sum()
    COP_VOLUME_NET_SPACE_BELOW_GROUND=space_df[is_net_space_below_ground]['NetVolume'].sum()

    BUILDING_VOLUME=COP_VOLUME_NET_SPACE+COP_KV
    #BUILDING_VOLUME_ABOVE_GROUND=COP_VOLUME_NET_SPACE_ABOVE_GROUND + COP_KV_ABOVE_GROUND
    #BUILDING_VOLUME_BELOW_GROUND=COP_VOLUME_NET_SPACE_BELOW_GROUND + COP_KV_BELOW_GROUND

    kv_dict = {
        'COP_KV': [COP_KV],
        #'COP_KV_ABOVE_GROUND': [COP_KV_ABOVE_GROUND],
        #'COP_KV_BELOW_GROUND': [COP_KV_BELOW_GROUND],
        #'COP_VOLUME_NET_SPACE': [COP_VOLUME_NET_SPACE],
        #'COP_VOLUME_NET_SPACE_ABOVE_GROUND': [COP_VOLUME_NET_SPACE_ABOVE_GROUND],
        #'COP_VOLUME_NET_SPACE_BELOW_GROUND': [COP_VOLUME_NET_SPACE_BELOW_GROUND],
        'BUILDING_VOLUME': [BUILDING_VOLUME],
        #'BUILDING_VOLUME_ABOVE_GROUND': [BUILDING_VOLUME_ABOVE_GROUND],
       # 'BUILDING_VOLUME_BELOW_GROUND': [BUILDING_VOLUME_BELOW_GROUND],
    }

    df_return = pd.DataFrame(kv_dict)

    return(df_return)

df_general_volume_test = general_volume_metrics(df_test_data)
df_general_volume_test.head()
```

```python

def calculate_facade_metrics(df: pd.DataFrame) -> pd.DataFrame:
    """
    Given a raw DataFrame (compliant with our format), returns the facade metrics.
    """

    # Filter dataframes for coverings and windows
    coverings_df = df[df['IfcElement'] == 'COVERING'].copy()
    windows_df = df[df['IfcElement'] == 'WINDOW'].copy()
    doors_df = df[df['IfcElement'] == 'DOOR'].copy()


    # Define conditions for different facade elements
    is_facade = (coverings_df['PredefinedType'] == 'CLADDING') & (coverings_df['IsExternal'] == 1)
    is_windows_external = windows_df['IsExternal'] == 1
    is_door_external = doors_df['IsExternal'] == 1
    is_roofing = (coverings_df['PredefinedType'] == 'ROOFING')
    is_insulation = (coverings_df['PredefinedType'] == 'INSULATION')

    # Calculate quantities
    AREA_CLADDING_EXTERNAL = coverings_df[is_facade]['NetArea'].sum()
    AREA_CLADDING_EXTERNAL_ABOVE_GROUND = coverings_df[above_ground(coverings_df, is_facade)]['NetArea'].sum()
    AREA_CLADDING_EXTERNAL_BELOW_GROUND = coverings_df[below_ground(coverings_df, is_facade)]['NetArea'].sum()

    AREA_TOTAL_ROOF = coverings_df[is_roofing]['NetArea'].sum()
    AREA_TOTAL_ROOF_ABOVE_GROUND = coverings_df[above_ground(coverings_df, is_roofing)]['NetArea'].sum()
    AREA_TOTAL_ROOF_BELOW_GROUND = coverings_df[below_ground(coverings_df, is_roofing)]['NetArea'].sum()

    AREA_TOTAL_INSULATION = coverings_df[is_insulation]['NetArea'].sum()
    AREA_TOTAL_INSULATION_ABOVE_GROUND = coverings_df[above_ground(coverings_df, is_insulation)]['NetArea'].sum()
    AREA_TOTAL_INSULATION_BELOW_GROUND = coverings_df[below_ground(coverings_df, is_insulation)]['NetArea'].sum()

    AREA_WINDOW_EXTERNAL = windows_df[is_windows_external]['Area'].sum()
    AREA_WINDOW_EXTERNAL_ABOVE_GROUND = windows_df[above_ground(windows_df, is_windows_external)]['Area'].sum()
    AREA_WINDOW_EXTERNAL_BELOW_GROUND = windows_df[below_ground(windows_df, is_windows_external)]['Area'].sum()

    AREA_DOOR_EXTERNAL = doors_df[is_door_external]['Area'].sum()
    AREA_DOOR_EXTERNAL_ABOVE_GROUND = doors_df[above_ground(doors_df, is_door_external)]['Area'].sum()
    AREA_DOOR_EXTERNAL_BELOW_GROUND = doors_df[below_ground(doors_df, is_door_external)]['Area'].sum()

    AREA_TOTAL_FACADE = AREA_CLADDING_EXTERNAL + AREA_WINDOW_EXTERNAL + AREA_DOOR_EXTERNAL
    AREA_TOTAL_FACADE_ABOVE_GROUND = AREA_CLADDING_EXTERNAL_ABOVE_GROUND + AREA_WINDOW_EXTERNAL_ABOVE_GROUND + AREA_DOOR_EXTERNAL_ABOVE_GROUND
    AREA_TOTAL_FACADE_BELOW_GROUND = AREA_CLADDING_EXTERNAL_BELOW_GROUND + AREA_WINDOW_EXTERNAL_BELOW_GROUND + AREA_DOOR_EXTERNAL_BELOW_GROUND

    AREA_TOTAL_HULL = AREA_CLADDING_EXTERNAL + AREA_WINDOW_EXTERNAL + AREA_TOTAL_ROOF + AREA_TOTAL_INSULATION
    AREA_TOTAL_HULL_ABOVE_GROUND = AREA_CLADDING_EXTERNAL_ABOVE_GROUND + AREA_WINDOW_EXTERNAL_ABOVE_GROUND + AREA_TOTAL_ROOF_ABOVE_GROUND + AREA_TOTAL_INSULATION_ABOVE_GROUND
    AREA_TOTAL_HULL_BELOW_GROUND = AREA_CLADDING_EXTERNAL_BELOW_GROUND + AREA_WINDOW_EXTERNAL_BELOW_GROUND + AREA_TOTAL_ROOF_BELOW_GROUND + AREA_TOTAL_INSULATION_BELOW_GROUND

    # Create dictionary for DataFrame construction
    facade_dict = {
        'AREA_CLADDING_EXTERNAL': [AREA_CLADDING_EXTERNAL],
        'AREA_CLADDING_EXTERNAL_ABOVE_GROUND': [AREA_CLADDING_EXTERNAL_ABOVE_GROUND],
        'AREA_CLADDING_EXTERNAL_BELOW_GROUND': [AREA_CLADDING_EXTERNAL_BELOW_GROUND],
        'AREA_TOTAL_FACADE': [AREA_TOTAL_FACADE],
        'AREA_TOTAL_FACADE_ABOVE_GROUND': [AREA_TOTAL_FACADE_ABOVE_GROUND],
        'AREA_TOTAL_FACADE_BELOW_GROUND': [AREA_TOTAL_FACADE_BELOW_GROUND],
        'AREA_TOTAL_ROOF': [AREA_TOTAL_ROOF],
        'AREA_TOTAL_ROOF_ABOVE_GROUND': [AREA_TOTAL_ROOF_ABOVE_GROUND],
        'AREA_TOTAL_ROOF_BELOW_GROUND': [AREA_TOTAL_ROOF_BELOW_GROUND],
        'AREA_WINDOW_EXTERNAL': [AREA_WINDOW_EXTERNAL],
        'AREA_WINDOW_EXTERNAL_ABOVE_GROUND': [AREA_WINDOW_EXTERNAL_ABOVE_GROUND],
        'AREA_WINDOW_EXTERNAL_BELOW_GROUND': [AREA_WINDOW_EXTERNAL_BELOW_GROUND],
        'AREA_TOTAL_INSULATION' : [AREA_TOTAL_INSULATION],
        'AREA_TOTAL_INSULATION_ABOVE_GROUND' : [AREA_TOTAL_INSULATION_ABOVE_GROUND],
        'AREA_TOTAL_INSULATION_BELOW_GROUND' : [AREA_TOTAL_INSULATION_BELOW_GROUND],
        'AREA_TOTAL_HULL': [AREA_TOTAL_HULL],
        'AREA_TOTAL_HULL_ABOVE_GROUND': [AREA_TOTAL_HULL_ABOVE_GROUND],
        'AREA_TOTAL_HULL_BELOW_GROUND': [AREA_TOTAL_HULL_BELOW_GROUND],
    }

    # Construct DataFrame
    df_return = pd.DataFrame(facade_dict)

    return df_return

df_facade_test = calculate_facade_metrics(df_test_data)
df_facade_test.head()
```

```python
def sia380_metrics(df: pd.DataFrame) -> pd.DataFrame:
  #TODO to calculate the "Thermische Gebäudehülle" the slabs need to be sperated at every border of the Space.

  spaces_df = df[df['IfcElement'] == 'SPACE'].copy()
  walls_df = df[df['IfcElement'] == 'WALLSTANDARDCASE'].copy()
  is_sia_ebf_room_max = spaces_df['EbfSia380Max'] == True
  is_sia_ebf_room_min = spaces_df['EbfSia380Min'] == True
  is_sia_ebf_wall_max = walls_df['EbfSia380Max'] == True
  is_sia_ebf_wall_min = walls_df['EbfSia380Min'] == True

  # Calculate Areas
  EBF_SPACE_AREA_MAX = spaces_df[is_sia_ebf_room_max]['NetFloorArea'].sum()
  EBF_SPACE_AREA_MIN = spaces_df[is_sia_ebf_room_min]['NetFloorArea'].sum()

  EBF_SPACE_VOLUME_MAX = spaces_df[is_sia_ebf_room_max]['NetVolume'].sum()
  EBF_SPACE_VOLUME_MIN = spaces_df[is_sia_ebf_room_min]['NetVolume'].sum()

  EBF_WALL_AREA_MAX = walls_df[is_sia_ebf_wall_max].apply(lambda row: row['Length'] * row['Width'], axis=1).sum()
  EBF_WALL_AREA_MIN = walls_df[is_sia_ebf_wall_min].apply(lambda row: row['Length'] * row['Width'], axis=1).sum()

  EBF_WALL_VOLUME_MAX = walls_df[is_sia_ebf_wall_max].apply(lambda row: row['Length'] * row['Width']* row['Height'], axis=1).sum()
  EBF_WALL_VOLUME_MIN = walls_df[is_sia_ebf_wall_min].apply(lambda row: row['Length'] * row['Width']* row['Height'], axis=1).sum()

  EBF_AREA_MAX = EBF_SPACE_AREA_MAX + EBF_WALL_AREA_MAX
  EBF_AREA_MIN = EBF_SPACE_AREA_MIN + EBF_WALL_AREA_MIN

  EBF_VOLUME_MAX = EBF_SPACE_VOLUME_MAX + EBF_WALL_VOLUME_MAX
  EBF_VOLUME_MIN = EBF_SPACE_VOLUME_MIN + EBF_WALL_VOLUME_MIN

  sia180_dict = {
      'EBF_AREA_MAX': EBF_AREA_MAX,
      'EBF_AREA_MIN': EBF_AREA_MIN,
      'EBF_VOLUME_MAX': EBF_VOLUME_MAX,
      'EBF_VOLUME_MIN': EBF_VOLUME_MIN,
  }

  df_return = pd.DataFrame(sia180_dict, index=[0])  # Creating a DataFrame with a single row
  return df_return

df_sia380_test = sia380_metrics(df_test_data)
df_sia380_test.head()
```

```python
def write_df_to_excel_sheet(excel_file: IO, df: pd.DataFrame, sheet_name: str) -> IO:
  with pd.ExcelWriter(excel_file, engine='openpyxl', mode='a') as writer:
    #TODO why is this necessary?
    df = df.drop_duplicates(subset=['Guid'])
    df.to_excel(writer, sheet_name=sheet_name, index=False)

#Assuming excel_file and df_test_data are defined somewhere before calling the function
#write_df_to_excel_sheet(test, df_test_data, 'test')
```

```python
def append_to_worksheet(excel_file: IO, sheet_name: str, df: pd.DataFrame)-> IO:

    wb = load_workbook(excel_file)

    ws = wb[sheet_name]
    next_row = ws.max_row + 1
    for index, row in df.iterrows():
        for col_idx, value in enumerate(row, start=1):
            ws.cell(row=next_row + index, column=col_idx, value=value)
    wb.save(excel_file)
```
